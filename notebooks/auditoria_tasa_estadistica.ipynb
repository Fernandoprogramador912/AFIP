```python
import sqlite3
from pathlib import Path

root = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
db = root / "data" / "tasa_estadistica.db"
print("DB:", db.resolve(), "exists:", db.is_file())

if db.is_file():
    con = sqlite3.connect(db)
    con.row_factory = sqlite3.Row
    cur = con.cursor()
    cur.execute("SELECT name FROM sqlite_master WHERE type='table' ORDER BY name")
    print("Tablas:", [r[0] for r in cur.fetchall()])
    cur.execute(
        """SELECT c.codigo, c.descripcion, c.importe, c.es_tasa_estadistica, c.match_reason
           FROM conceptos_liquidacion c ORDER BY c.id DESC LIMIT 20"""
    )
    for row in cur.fetchall():
        print(dict(row))
    con.close()
```
